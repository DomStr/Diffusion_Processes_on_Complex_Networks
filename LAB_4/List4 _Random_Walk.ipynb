```python
import numpy as np
import random
import imageio
import matplotlib.pyplot as plt
import os
import pygame
import networkx as nx

def random_walk(n, size):
    
    x = np.zeros(n) #stating number of moves, starting point:(0,0)
    y = np.zeros(n)
    
    X = size[0] #grid size
    Y = size[1]

    for i in range(1, n): 
        if x[i-1] > X-1: #We are on the right edge - we have 3 possible directions
            if y[i-1] > Y-1: #We are in the righ up corner - we have 2 possible directions
                direction = random.randint(1, 2) #randomly chosing one from 2 directions
                if direction == 1:
                    x[i] = x[i - 1] - 1 #lets go left
                    y[i] = y[i - 1]
                else:
                    x[i] = x[i - 1]     #lets go down
                    y[i] = y[i - 1] - 1
            elif y[i-1] < -Y+1: #We are in the right down corner
                direction = random.randint(1, 2)
                if direction == 1:
                    x[i] = x[i - 1] - 1 #lets go left
                    y[i] = y[i - 1]
                else:
                    x[i] = x[i - 1]     #lets go up
                    y[i] = y[i - 1] + 1         
                    
            else: #We are on right edge
                direction = random.randint(1, 3)
                if direction == 1:
                    x[i] = x[i - 1] - 1 #lets go left
                    y[i] = y[i - 1]
                elif direction == 2:
                    x[i] = x[i - 1]     #lets go up
                    y[i] = y[i - 1] + 1
                else:
                    x[i] = x[i - 1]     #lets go down
                    y[i] = y[i - 1] - 1
        
        elif x[i-1] < -X+1:
            if y[i-1] > Y-1: #We are in the left up corner
                direction = random.randint(1, 2) 
                if direction == 1:
                    x[i] = x[i - 1] + 1 #lets go right
                    y[i] = y[i - 1]
                else:
                    x[i] = x[i - 1]     #lets go down
                    y[i] = y[i - 1] - 1
            elif y[i-1] < -Y+1: #We are in the left down corner
                direction = random.randint(1, 2)
                if direction == 1:
                    x[i] = x[i - 1] + 1 #lets go right
                    y[i] = y[i - 1]
                else:
                    x[i] = x[i - 1]     #lets go up
                    y[i] = y[i - 1] + 1         
                    
            else: #We are on left edge
                direction = random.randint(1, 3)
                if direction == 1:
                    x[i] = x[i - 1] + 1 #lets go right
                    y[i] = y[i - 1]
                elif direction == 2:
                    x[i] = x[i - 1]     #lets go up
                    y[i] = y[i - 1] + 1
                else:
                    x[i] = x[i - 1]     #lets go down
                    y[i] = y[i - 1] - 1
        
            
            
        else: #We are not on the edge and in the corner
            direction = random.randint(1, 4)
            if direction == 1:
                x[i] = x[i - 1] + 1 #lets go right
                y[i] = y[i - 1]
            elif direction == 2:
                x[i] = x[i - 1] - 1 #lets go left
                y[i] = y[i - 1]
            elif direction == 3:
                x[i] = x[i - 1]     #lets go up
                y[i] = y[i - 1] + 1
            else:
                x[i] = x[i - 1]     #lets go down
                y[i] = y[i - 1] - 1
            
    
    plt.figure(figsize=(X+1, Y+1)) #grid in size of (X,Y)
    images = [] #empty set of images
    plt.grid()
    plt.xlim(-X-1,X+1)
    plt.ylim(-Y-1,Y+1)
    
    for j in range(1,n): #saving image after each step
        plt.plot(x[j-1:j+1], y[j-1:j+1],linestyle='-',marker='o', color='black')
        plt.savefig('C:/Programy/python/Python 3.8/images/' + str(2*j) + '.png')
        images.append(imageio.imread('C:/Programy/python/Python 3.8/images/' + str(2*j) + '.png'))
        plt.plot(x[j],y[j],marker='o',color='red')
        plt.savefig('C:/Programy/python/Python 3.8/images/' + str((2*j)+1) + '.png')
        images.append(imageio.imread('C:/Programy/python/Python 3.8/images/' + str((2*j)+1) + '.png'))

    imageio.mimwrite("images/animation.avi", images[:], fps=6)
```

```python
def findAnBn(n): #We are looking for points in our random walk which are on the right half or in the first quadrant 
    alpha=2*np.pi*np.random.rand(n)
    x=np.cumsum(np.sin(alpha))
    y=np.cumsum(np.cos(alpha))
    An=np.sum(x>0)/n
    Bn=np.sum((x>0) & (y>0))/n
    return An,Bn
```

```python
An,Bn = zip(*[findAnBn(n) for _ in range(10000)])
```

```python
MeanAn = np.mean(An)
MeanBn = np.mean(Bn)
print('Expected value that walker is in the right half', MeanAn)
print('Expected value that walker is in the first quadrant', MeanBn)
```
